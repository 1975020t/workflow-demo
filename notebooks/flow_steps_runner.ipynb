```python
from src.agent import HelpDeskWorkflow, WorkflowState
from src.configs import Settings
```

```python
settings = Settings()
workflow = HelpDeskWorkflow(settings=settings)
```

```python
question = "パスワードの文字数制限と使える特殊文字を教えてください"
```

## Step 1: 質問分析（キーワード抽出 + ルーティング判定）

```python
state: WorkflowState = {
    "question": question,
    "analysis": None,
    "search_results": [],
    "filtered_data": [],
    "answer": "",
}

step1_result = workflow.analyze_question(state)
state.update(step1_result)
```

```python
print(f"データソース: {state['analysis'].data_source}")
print(f"キーワード:   {state['analysis'].keywords}")
```

## Step 2: データ検索（プログラム的 - LLM不要）

```python
step2_result = workflow.search_data(state)
state.update(step2_result)
```

```python
print(f"検索結果: {len(state['search_results'])}件\n")
for i, r in enumerate(state["search_results"]):
    print(f"[{i+1}] ({r.file_name})")
    print(f"    {r.content[:100]}...")
    print()
```

```python
step3_result = workflow.filter_results(state)
state.update(step3_result)
```

```python
print(f"絞り込み結果: {len(state['filtered_data'])}件\n")
for i, entry in enumerate(state["filtered_data"]):
    print(f"[{i+1}] {entry}")
    print()
```

## Step 4: 回答生成（LLM）

```python
step4_result = workflow.generate_answer(state)
state.update(step4_result)
```

```python
print("=" * 50)
print("最終回答")
print("=" * 50)
print(state["answer"])
```
